### Using Escherichia coli K-12 MG1655 on glucose minimal media at 37°C

By keeping E. coli in constant substrate excess and exponential growth, fitness increases up to 1.6-fold were obtained compared to the wild type. These increases are comparable to previously reported maximum growth rates in similar conditions but were obtained over a shorter time frame.

```python
# GLU with snp mutation within -10box

GLU_within_box_10 = []
for i in range(len(mapped_features_10_GLU)):
    if mapped_features_10_GLU.box_10_left.iloc[i] <= mapped_features_10_GLU.snp_position.iloc[i] <= mapped_features_10_GLU.box_10_right.iloc[i]:
        GLU_within_box_10.append(mapped_features_10_GLU.snp_position.iloc[i])

len(GLU_within_box_10)
```
